# 2. Given an unsorted array of integers, find the length of the longest continuous


```python
def find_length(nums):
    if not nums:
        return 0
    
    max_length = 1
    current_length = 1
    
    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_length += 1
        else:
            max_length = max(max_length, current_length)
            current_length = 1
    
    return max(max_length, current_length)

# Example 
nums = [1, 3, 5, 4, 7]
print(find_length(nums))
```

# 3. Given a list of non negative integers, arrange them such that they form the largest


```python
def largest_number(nums):
    nums = [str(num) for num in nums]
    nums.sort(key=lambda x: x*10, reverse=True)
    return str(int("".join(nums)))

# Example 
nums = [10,2]
print(largest_number(nums))  # Output: "9534330"
```
